```python
import pandas as pd
```

```python
df= pd.read_csv("house_prices.csv.zip")
```

```python
df.columns
```

```python
df = df[['Carpet Area', 'Price (in rupees)']]
df.head()
```

```python
df['Carpet Area'] = df['Carpet Area'].str.replace(' sqft', '', regex=False)
df['Carpet Area'] = pd.to_numeric(df['Carpet Area'], errors='coerce')
```

```python
df.dropna(inplace=True)
```

```python
df.columns = ['Area', 'Price']
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv("clean_house_data.csv", index=False)
```
